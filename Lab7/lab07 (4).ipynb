Діедральна група $D_{20}$ має ряд нормальних підгруп: 

```python
from sage.groups.perm_gps.permgroup_named import DihedralGroup
G = DihedralGroup(20)
normal_subgroups = G.normal_subgroups()
print("Усього нормальних підгруп:", len(normal_subgroups))
for i, N in enumerate(normal_subgroups):
    print(f"Номер: {i + 1}, Порядок: {N.order()}, Генератори: {N.gens()}")
print("-" * 50)

for i, N in enumerate(normal_subgroups):
    images = []
    for g in G.gens():
        if g in N:
            images.append(G.identity()) 
        else:
            images.append(g)

    phi = G.hom(images, codomain=G)
    kernel = phi.kernel()
    print(f"Номер підгрупи: {i + 1}")
    print(f"Підгрупа N (порядок {N.order()}): {N}")
    print(f"Ядро φ: {kernel}")
    print(f"Ядро збігається з N: {kernel == N}")
    print("-" * 50)
```

```python
G4 = graphs.CubeGraph(4)
G4
```

```python

Aut_G4 = G4.automorphism_group()
print("\nГрупа автоморфізмів графа G4:")
print("Порядок групи:", Aut_G4.order())


orbits = Aut_G4.orbits()
print("\nОрбіти дії групи автоморфізмів на вершинах:")
for i, orbit in enumerate(orbits):
    print(f"Орбіта {i+1}: {orbit}")

vertex = G4.vertices()[0] 
stabilizer = Aut_G4.stabilizer(vertex)
print("\nСтабілізатор вершини 0:")
print("Порядок стабілізатора:", stabilizer.order())


stabilizer_action_orbits = stabilizer.orbits()
print("\nОрбіти дії стабілізатора вершини 0 на вершинах:")
for i, orbit in enumerate(stabilizer_action_orbits):
    print(f"Орбіта {i+1}: {orbit}")

# 6. Висновки
print("\nВисновки:")
print("1. Група автоморфізмів графа G4 діє транзитивно на вершинах, тому всі вершини мають одну орбіту.")
print("2. Стабілізатор вершини є підгрупою автоморфізмів, і його дія на вершини визначає класи еквівалентності вершин за симетрією відносно заданої вершини.")
```
