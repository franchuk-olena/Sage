## –ó–∞–≤–¥–∞–Ω–Ω—è 1

–í –≥—Ä—É–ø—ñ —Ä—É—Ö—ñ–≤ –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –≤–æ—Å—å–º–∏–∫—É—Ç–Ω–∏–∫–∞ –∑–∞–¥–∞–π—Ç–µ —Ü–∏–∫–ª—ñ—á–Ω—É –ø—ñ–¥–≥—Ä—É–ø—É –ø–æ—Ä—è–¥–∫—É 4. –ü–æ–±—É–¥—É–π—Ç–µ –¥–ª—è –Ω–µ—ó –ª—ñ–≤—ñ —ñ –ø—Ä–∞–≤—ñ –∫–ª–∞—Å–∏ —Å—É–º—ñ–∂–Ω–æ—Å—Ç—ñ —Ç–∞ –ø–µ—Ä–µ–≤—ñ—Ä—Ç–µ —á–∏ –±—É–¥–µ —Ü—è –≥—Ä—É–ø–∞ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—é. 

```python
D8 = DihedralGroup(8)
a = D8('(1,3,5,7)(2,4,6,8)')
H = D8.subgroup([a])
print(H.order())

#–ø—Ä–∞–≤—ñ —ñ –ª—ñ–≤—ñ –∫–ª–∞—Å–∏ —Å—É–º—ñ–∂–Ω–æ—Å—Ç—ñ
print("Right cosets:")
right_cosets = D8.cosets(H, side='right')
print(right_cosets)

print("Left cosets:")
left_cosets = D8.cosets(H, side='left')
print(left_cosets)

sorted_right_cosets = sorted([sorted(coset) for coset in right_cosets])
sorted_left_cosets = sorted([sorted(coset) for coset in left_cosets])

is_normal = sorted_right_cosets == sorted_left_cosets
print("Is the subgroup normal?", is_normal)
```

## –ó–∞–≤–¥–∞–Ω–Ω—è 2 

–ü–µ—Ä–µ–±–µ—Ä—ñ—Ç—å –≤—Å—ñ –ø—ñ–¥–≥—Ä—É–ø–∏ –∑–Ω–∞–∫–æ–∑–º—ñ–Ω–Ω–æ—ó –≥—Ä—É–ø–∏ $A_5$. –ü–µ—Ä–µ–≤—ñ—Ä—Ç–µ –≤—Å—ñ –Ω–∞ –Ω–æ—Ä–º–∞–ª—å–Ω—ñ—Å—Ç—å. –ü–æ—Ä—ñ–≤–Ω—è–π—Ç–µ —á–∞—Å —Ä–æ–±–æ—Ç–∏ –∑ –º–µ—Ç–æ–¥–æ–º `.is_simple` –¥–ª—è –≥—Ä—É–ø–∏ $A_5$. 

```python
import time
from sage.groups.perm_gps.permgroup_named import AlternatingGroup

A5 = AlternatingGroup(5)
start_time = time.time()

#–ø–µ—Ä–µ–≤—ñ—Ä–∫–∞ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç—ñ –ø—ñ–¥–≥—Ä—É–ø–∏
def is_normal(subgroup, group) -> bool: 
    return group.cosets(subgroup, side='right') == group.cosets(subgroup, side='left')

#–ø–µ—Ä–µ–≤—ñ—Ä–∫–∞ –ø—Ä–æ—Å—Ç–æ—Ç–∏ –≥—Ä—É–ø–∏
def simple_check(group) -> bool:
    trivial_subgroup = group.subgroups()[0]  # –¢—Ä–∏–≤—ñ–∞–ª—å–Ω–∞ –ø—ñ–¥–≥—Ä—É–ø–∞
    for subgroup in group.subgroups():
        if subgroup.order() not in [1, group.order()] and is_normal(subgroup, group):
            return False 
    return True

print(simple_check(A5))
end_time = time.time()
print(f"My method {end_time - start_time:.4f} seconds")


start_time = time.time()
print(A5.is_simple())
end_time = time.time()
print(f"Built-in method {end_time - start_time:.4f} seconds")
```

–°–ø—Ä–æ–≥–Ω–æ–∑—É–π—Ç–µ —Å–∫—ñ–ª—å–∫–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–∏—Ö –ø—ñ–¥–≥—Ä—É–ø –º–∞—î –≥—Ä—É–ø–∞ $D_{470448}$ __–±–µ–∑ –≤–∏–≤–æ–¥—É –≤—Å—ñ—Ö –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é Sage__. _–û–ø–∏—à—ñ—Ç—å_ –≤—Å—ñ –Ω–æ—Ä–º–∞–ª—å–Ω—ñ –ø—ñ–¥–≥—Ä—É–ø–∏ –≥—Ä—É–ø–∏ $D_{470448}$.  

```python
from sage.groups.perm_gps.permgroup_named import DihedralGroup

selected_n = [4, 25, 70]

normal_subgroups_orders = {}

for n in selected_n:
    D_n = DihedralGroup(n)
    normal_subgroups = D_n.normal_subgroups()
    orders = [H.order() for H in normal_subgroups]
    normal_subgroups_orders[n] = orders
for n, orders in normal_subgroups_orders.items():
    print(f"D{n}: Normal subgroup orders = {orders}")
```
